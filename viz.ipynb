## categorical vs categorical

```python
top3_pub = data['Publisher'].value_counts().index[:7]
top3_gen = data['Genre'].value_counts().index[:7]
top3_plat = data['Platform'].value_counts().index[:7]
top3_data = data.loc[(data['Publisher'].isin(top3_pub)) & (data['Platform'].isin(top3_plat)) & (data['Genre'].isin(top3_gen))]
top3_data

#### to comapre between data
```

```python
plt.figure(figsize=(10,9))
sns.countplot(x='Publisher', hue='Platform', data=top3_data)
plt.xlabel('Publisher')
plt.ylabel('Count of Games')
plt.show()
```

```python
sns.boxplot(x='Publisher',y='EU_Sales',data=top3_data)
plt.xticks(rotation=90 ,  fontsize=15)
plt.yticks(fontsize=12)
plt.title('sales for top3 publisher',fontsize=15)
plt.show

### to see the boxplots between datas
```

```python
sns.barplot(data=top3_data, x='Genre', y='Global_Sales')
```

```python
plt.figure(figsize=(20,10))
plt.suptitle('video games sales dashboard',fontsize=20)


plt.subplot(2,3,1)
sns.scatterplot(x='EU_Sales',y='Global_Sales', color ='red' , data=top3_data)
plt.title('EU Sales & Global Sales',fontsize=16)
plt.xlabel('EU Sales', fontsize=12)
plt.ylabel('Global',fontsize=12)


plt.subplot(2,3,2)
sns.scatterplot(x='NA_Sales',y='Global_Sales',color ='blue', data=top3_data)
plt.title('NA Sales & Global Sales',fontsize=16)
plt.xlabel('NA Sales', fontsize=12)
plt.ylabel('Global',fontsize=12)


plt.subplot(2,3,3)
sns.scatterplot(x='JP_Sales',y='Global_Sales',color = 'green', data=top3_data)
plt.title('JP Sales & Global Sales',fontsize=16)
plt.xlabel('JP sales', fontsize=12)
plt.ylabel('Global',fontsize=12)

plt.subplot(2,3,4)
sns.scatterplot(x='Other_Sales',y='Global_Sales',color = 'black', data=top3_data)
plt.title('Other & Global Sales',fontsize=16)
plt.xlabel('Other sales', fontsize=12)
plt.ylabel('Global',fontsize=12)


plt.tight_layout(rect=[0,0,1,0.95])
plt.show()


```
